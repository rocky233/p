```python
plt.figure(figsize=(10, 6))
male_age=df[df['Sex']=='male']['Age'].dropna()
female_age = df[df['Sex'] == 'female']['Age'].dropna()
plt.boxplot([male_age, female_age], labels=['Male', 'Female'])
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Distribution of Age by Gender')
plt.show()
```

```python
sns.boxplot(x='Sex',y='Age',data=df,hue='Survived')
```

```python
# Observations
male_age = df[(df['Sex'] == 'male') & (df['Age'].notnull())]['Age']
female_age = df[(df['Sex'] == 'female') & (df['Age'].notnull())]['Age']
print(male_age)
print(female_age)
```

```python
print('Observations:')
print(f'Median age for males: {male_age.median()}')
print(f'Median age for females: {female_age.median()}')
print(f'Number of males: {male_age.count()}')
print(f'Number of females: {female_age.count()}')
print(f'Number of males survived: {df[(df["Sex"] == "male") & (df["Survived"] == 1)]["Age"].count()}')
print(f'Number of females survived: {df[(df["Sex"] == "female") & (df["Survived"] == 1)]["Age"].count()}')
```
